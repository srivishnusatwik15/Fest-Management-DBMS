```python
with open('o.txt', 'r') as f:
    lines = f.readlines()

data = [[]]
for line in lines:
    if any(c.isalpha() for c in line) or not line.strip():
        data.append([])
    else:
        data.append([float(x) for x in line.split()])

# Now data[i] contains the numeric values from each non-alphabetic and non-empty line in the file




```

```python
import matplotlib.pyplot as plt
diskReadsLRU = data[8]
diskReadsMRU = data[23]
diskReadsClock = data[38]

frames = [i for i in range(2, 21)]

plt.plot(frames[::-1], diskReadsLRU[-19:], label='LRU')
plt.plot(frames[::-1], diskReadsMRU[-19:], label='MRU')
plt.plot(frames[::-1], diskReadsClock[-19:], label='Clock')
plt.scatter(frames[::-1], diskReadsLRU[-19:])
plt.scatter(frames[::-1], diskReadsMRU[-19:])
plt.scatter(frames[::-1], diskReadsClock[-19:])
plt.xlabel('Number of Frames in Buffer Pool')
plt.ylabel('Number of Disk Reads')
plt.legend()
plt.title('Disk Reads for Join Query ')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Function to calculate average time
def calculate_average_time(data):
    total_time = sum(data)
    count = len(data)
    if count > 0:
        return total_time / count
    else:
        return 0

# Extracting time data for each query type
lru_select_time = data[4]
lru_join_time = data[9]
lru_selfjoin_time = 1000*data[14]

mru_select_time = data[19]
mru_join_time = data[24]
mru_selfjoin_time = 1000*data[29]

clock_select_time = data[34]
clock_join_time = data[39]
clock_selfjoin_time = 1000*data[44]

# Calculating average times for each query type
avg_lru_select_time = calculate_average_time(lru_select_time)
avg_lru_join_time = calculate_average_time(lru_join_time)
avg_lru_selfjoin_time = calculate_average_time(lru_selfjoin_time)

avg_mru_select_time = calculate_average_time(mru_select_time)
avg_mru_join_time = calculate_average_time(mru_join_time)
avg_mru_selfjoin_time = calculate_average_time(mru_selfjoin_time)

avg_clock_select_time = calculate_average_time(clock_select_time)
avg_clock_join_time = calculate_average_time(clock_join_time)
avg_clock_selfjoin_time = calculate_average_time(clock_selfjoin_time)

# Data for visualization
query_types = ['LRU Select', 'LRU Join', 'LRU Self-Join',
               'MRU Select', 'MRU Join', 'MRU Self-Join',
               'Clock Select', 'Clock Join', 'Clock Self-Join']
avg_times = [avg_lru_select_time, avg_lru_join_time, avg_lru_selfjoin_time,
             avg_mru_select_time, avg_mru_join_time, avg_mru_selfjoin_time,
             avg_clock_select_time, avg_clock_join_time, avg_clock_selfjoin_time]
print("Average time taken for LRU select:", avg_lru_select_time)
print("Average time taken for LRU join:", avg_lru_join_time)
print("Average time taken for LRU self-join:", avg_lru_selfjoin_time)

print("Average time taken for MRU select:", avg_mru_select_time)
print("Average time taken for MRU join:", avg_mru_join_time)
print("Average time taken for MRU self-join:", avg_mru_selfjoin_time)

print("Average time taken for CLOCK select:", avg_clock_select_time)
print("Average time taken for CLOCK join:", avg_clock_join_time)
print("Average time taken for CLOCK self-join:", avg_clock_selfjoin_time)

# Visualizing average times using a bar chart
plt.figure(figsize=(10, 6))
plt.barh(query_types, avg_times, color='skyblue')
plt.xlabel('Average Time (s)')
plt.title('Average Time Taken for Different Query Types')
plt.gca().invert_yaxis()  # Invert y-axis to display query types from top to bottom
plt.show()
```
